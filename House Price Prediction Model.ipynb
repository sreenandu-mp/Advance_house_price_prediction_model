```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("X_train.csv")
dataset.shape
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset.drop(columns=['SalePrice']),dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
#making a predicted y value for given X
y_pred=reg.predict(X_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE:%.2f'%mean_squared_error(y_test,y_pred))
print('R2_score:%.2f'%r2_score(y_test,y_pred))
```
